# TimeSeries Index

* pd.to_datetime</br>
* pd.date_range</br>
* Year, Month, day 추출하는 방법</br>
* 월별 수익률을 구하는 방법</br>

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./Data/10Y Rate.csv', index_col= 'Date')
df.index

## index가 object 타입 datetime으로 바꿔줘야 한다! pd.to_datetime()

df.index = pd.to_datetime(df.index) # 아쉽게도 inplace는 파라미터에 없다
df.index
```

```python
# pd.date_range()

timeSeries_index = pd.date_range(start = '2020-10-25', end = None, periods = 100)
timeSeries_index
```

```python
# 년, 월, 일 구하기

df.index.year
df.index.month
df.index.day

df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day

df
```

### 10년 금리에서(국고채)에서 각 월별 수익률을 구하여라!

```python
df.groupby('Month').mean()[['Change']]
```

```python
df['Condition'] = np.where(df['Month'] == df['Month'].shift(1), 0 ,1)
df['indexNumber'] = range(df.shape[0])
index_condtion_number=df[df['Condition'] == 1]['indexNumber'].values -1 # 이게 연말의 인덱스값

df_condtion = df.iloc[index_condtion_number[1:]]
df_monthly_return=df_condtion['Close'].pct_change()
df_monthly_return

pd.DataFrame(df_monthly_return)
```
