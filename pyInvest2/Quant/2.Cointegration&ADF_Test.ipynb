## 1. Correlation & Cointgration

- 왜 pair trading 할 때 correlation이 높은 데이터 셋이 아닌, cointegrated한 데이터 셋을 뽑아야 할까?

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
def norm_dist_data(mu,sigma):
    return np.random.normal(mu,sigma)
```

```python
N = 1000
mu = 0
sigma = 10

random_variable_x = [norm_dist_data(mu + i,sigma) for i in range(N)] # non-stationary-data1
random_variable_y = [norm_dist_data(mu + i*2,sigma) for i in range(N)] # non-stationary-data2
data_index = [i for i in range(N)]

df_ex1 = pd.DataFrame([random_variable_x,random_variable_y]).T
df_ex1.index = data_index
df_ex1.columns = ['Random X','Random Y']

trace1 = go.Scatter(x=df_ex1.index,y=df_ex1['Random X'],mode='lines')
trace2 = go.Scatter(x=df_ex1.index,y=df_ex1['Random Y'],mode='lines')

layout = go.Layout(title='Random Variable')
fig = go.Figure([trace1,trace2], layout)
fig.show()
```

```python
df_ex1['Diff'] = df_ex1['Random X'] - df_ex1['Random Y']
trace3 = go.Scatter(x=df_ex1.index
                    ,y=df_ex1['Diff'],
                    mode='lines')
layout = go.Layout(title='EX1_Diff')
fig = go.Figure(trace3, layout)
fig.show()

## 여전히 추세 존재,,, -> 시간이 지나갈 수록 평균이 작아진다
```

## 2.ADF Test(단위근 검정)
- H0(귀무가설): 데이터에 단위근이 존재 -> Non-Stationary
- H1(대립가설): 시계열 데이터에 정상성을 만족 -> Stationary
- ADF Test 결과에서 P-Value가 신뢰수준보다 낮으면, 해당 시계열데이터는 Stationary라고 판단!
- https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html

```python
from statsmodels.tsa.stattools import adfuller
```

- df_ex1 = 두개의 Non-Stationary Data 선형조합 -> Non-Stationary
- df_ex2 = 두개의 Non-Stationary Data 선형조합 -> Stationary

```python
adfuller(df_ex1['Diff'])[1] < 0.01 # non-stationary
```

```python
adfuller(df_ex2['Diff'])[1] < 0.01 # stationary
```

```python
def adf_test(df, critical_value):
    p_value = adfuller(df)[1]
    
    if p_value < critical_value:
        print("Input TimeSeries Data is Stationary")
        return True
    else:
        print("Input TimeSeries Data is Not Stationary")
        return False
```

```python
adf_test(df = df_ex2['Diff'], critical_value = 0.01)
```

```python
adf_test(df = df_ex1['Diff'], critical_value = 0.01)
```
