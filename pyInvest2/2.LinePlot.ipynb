## Basic1

```python
import numpy as np
import plotly.graph_objs as go
import pandas as pd

np.random.seed(1)
random_x = np.random.randint(50, 100, 180)

datetime_index= pd.date_range('2022-05-17', '2022-11-12')

df = pd.DataFrame(random_x)
df.index = datetime_index
df.columns = ['Stock A']

trace=go.Scatter(x=df.index, y=df['Stock A'], mode='lines', name='Stock A')
layout = go.Layout(title='Stock A'
                    , xaxis= {'title':'datetime'}
                    , yaxis= {'title':'Stock A'})

fig = go.Figure(trace,layout)
fig.show()
```

```python
## 추세가 있는 데이터 만들기! -> 직선 + 추세 데이터
np.random.seed(1)
random_x = np.random.randint(50, 100, 180)
datetime_index = pd.date_range('2022-05-17','2022-11-12')

trend_factor = np.linspace(1, 180, 180)
timeSeries_data = random_x + trend_factor

df = pd.DataFrame(timeSeries_data)
df.index = datetime_index
df.columns = ['Stock A']


trace = go.Scatter(x = df.index, y = df['Stock A'], mode = 'lines', name = 'Stcok A')
layout = go.Layout(title = 'Stock A',
                  xaxis = {'title': 'datetime'},
                  yaxis = {'title': 'Stock A'})

fig = go.Figure(data = trace, layout = layout)
fig.show()
```

## Basic 2

```python
random_x1 = np.random.randint(50, 100, 180)
random_x2 = np.random.randint(50, 400, 180)

datetime_index = pd.date_range('2022-05-17', '2022-11-12')

trend_factor = np.linspace(1, 180, 180)
timeSeries_data1 = random_x1 + trend_factor
timeSeries_data2 = random_x2 + trend_factor

df = pd.DataFrame({'Stock A': timeSeries_data1, 'Stock B':timeSeries_data2})
df.index = datetime_index

trace1 = go.Scatter(x=df.index,y=df['Stock A'], mode='lines', name='Stock A')
trace2 = go.Scatter(x=df.index,y=df['Stock B'], mode='lines', name='Stock B')

layout = go.Layout(title= 'Stock A & Stock B', xaxis= {'title':'datetime'}, yaxis={'title': 'Stock'})
fig = go.Figure(data=[trace1,trace2],layout=layout) # 2개를 넣어줄 때에는 data에 리스트로 여러개의 trace를 넣어준다

fig.show()
```

```python
## 다음과 같은 형식으로 빈 리스트를 생성하고, append()를 통해서 만들어준다
data = []

trace1 = go.Scatter(x= df.index, y = df['Stock A'], mode = 'lines', name = 'Stock A')
data.append(trace1)
trace2 = go.Scatter(x = df.index, y = df['Stock B'], mode = 'lines', name = 'Stock B')
data.append(trace2)

layout = go.Layout(title = 'Stock A & Stock B', xaxis = {'title':'datetime'}, yaxis = {'title':'Stcok'})

fig = go.Figure(data = data, layout = layout)
fig.show()
```

```python
def cumreturn_line_plot(df, kind1, kind2, startDate, endDate):
    
    df_copy = df.copy()
    
    condition_kodex = [df_copy.columns[i] for i in range(df_copy.shape[1]) if 'KODEX' in df_copy.columns[i]]

    df_copy_kodex = df_copy.loc[:, condition_kodex]

    df_copy_kodex_set = df_copy_kodex.loc[:, [kind1, kind2]]
    df_copy_kodex_set = df_copy_kodex_set.dropna()

    startDate = startDate
    endDate = endDate

    df_copy_kodex_set_sample = df_copy_kodex_set.loc[startDate:endDate, :]

    df_copy_kodex_set_sample_return = ((df_copy_kodex_set_sample - df_copy_kodex_set_sample.iloc[0])/df_copy_kodex_set_sample.iloc[0])*100

    data = [go.Scatter(x = df_copy_kodex_set_sample_return.index,
                      y = df_copy_kodex_set_sample_return['{}'.format(i)],
                      mode = 'lines',
                      name = '{}'.format(i)) for i in df_copy_kodex_set_sample_return.columns]
    layout = go.Layout(title = '{} & {} cumReturn'.format(kind1, kind2))

    fig = go.Figure(data = data, layout = layout)
    fig.show()
```
