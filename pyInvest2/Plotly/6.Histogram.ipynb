## 1.Basic
- 히스토그램은 x축에 데이터를 넣어준다
- 금융공학 -> 주가 모델링 -> 상품 가격 매길 때 (가정)

```python
import numpy as np
import plotly.graph_objs as go

random_x = np.random.randint(0, 100, 100)

trace1 = go.Histogram(x = random_x, nbinsx = 100)

layout = go.Layout(title = 'Histogram of RandomVariable')

fig = go.Figure(data = trace1, layout = layout)
fig.show()
```

```python
import pandas as pd

df = pd.read_excel('./Data/df_etf.xlsx', index_col = 0)
df_copy = df.copy()
```

```python
# KODEX ETF Data

condition_kodex = [df_copy.columns[i] for i in range(df_copy.shape[1]) if 'KODEX' in df_copy.columns[i]]

df_copy_kodex = df_copy.loc[:, condition_kodex]

kind = 'KODEX 200'

# 가격 Data Histogram
df_copy_kodex_sample = df_copy_kodex.loc[:, [kind]]

trace1 = go.Histogram(x = df_copy_kodex_sample[kind],
                     xbins = dict(start = df_copy_kodex_sample[kind].min(),
                                 end = df_copy_kodex_sample[kind].max(),
                                 size = 100))

layout1 = go.Layout(title = 'Histogram of {} Price'.format(kind),
                   xaxis = {'tickformat':','})

fig1 = go.Figure(data = trace1, layout = layout1)
fig1.show()
```

```python
# 수익률 Data Histogram
df_copy_kodex_sample = df_copy_kodex.loc[:, [kind]]
df_copy_kodex_sample_return = df_copy_kodex_sample.pct_change(periods = 1).dropna()
df_copy_kodex_sample_return = np.round(df_copy_kodex_sample_return * 100, 1)
df_copy_kodex_sample_return
```

```python
trace2 = go.Histogram(x = df_copy_kodex_sample_return[kind],
                     xbins = dict(start = df_copy_kodex_sample_return[kind].min(),
                                 end = df_copy_kodex_sample_return[kind].max(),
                                 size =0.1))
layout2 = go.Layout(title = 'Histogram of {} Return'.format(kind))

fig2 = go.Figure(data = trace2, layout =layout2)
fig2.show()
```
